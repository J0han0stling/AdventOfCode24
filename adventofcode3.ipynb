```python
import re
```

```python
# Öppna och läs hela filen
with open("input_day3\\program.txt", "r", encoding="utf-8") as file:
    content = file.read()

# Visa innehållet
print(content) #lång lista. varje character är ett element i en lista
```

```python
# Regex-mönster
pattern = r"mul\(\d+,\d+\)"

# Hitta alla matchningar
muls = re.findall(pattern, content)

print(muls)
```

```python
pattern = r"\d+"

products = []
for mul in muls:
    integers = []
    integers.append(re.findall(pattern,mul)[0])
    integers.append(re.findall(pattern,mul)[1])
    products.append(int(integers[0])*int(integers[1]))
print(products)
print(sum(products))
    
```
